```
install.packages("ggplot2")
library(SportsAnalytics)
library(ggplot2)

NBA1314 <- fetch_NBAPlayerStatistics("13-14")
qplot(TotalMinutesPlayed, TotalPoints,data= NBA1314 )
#總上場時間(x軸)與總得分數(y軸)的關係
```

```
qplot(TotalMinutesPlayed, TotalPoints,data= NBA1314 ,geom=c("point","smooth"))
#總上場時間(x軸)與總得分數(y軸)的關係_漸進線
```

```
qplot(FieldGoalsMade, data= NBA1314 , fill = Position)
#投籃進球數(x軸)與守備位置(y軸)的關係
```

```
qplot(FieldGoalsAttempted,TotalPoints, data= NBA1314 , color = TotalPoints)+geom_smooth()
# 投籃出球數(x軸)與總得分數(y軸)的相關
```
